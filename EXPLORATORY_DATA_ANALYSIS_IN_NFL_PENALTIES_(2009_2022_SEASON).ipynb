9.Which  referee has called the most penalties in the NFL since 2009?

```python
# Group the data by referee and count the number of penalties called
referee_penalties = df.groupby('Ref_Crew')['Beneficiary'].count().reset_index()

# Sort the data in descending order to find the referee with the most penalties
referee_penalties = referee_penalties.sort_values(by='Beneficiary', ascending=False)

# Print the top referee
print(f"Referee with the most penalties since 2009: {referee_penalties.iloc[0]['Ref_Crew']}")
```

This will give you the referee Carl Cheffers who has called the most penalties in the NFL since 2009.

10.Which team has been penalized the most during this period?

```python
# Group the data by team and count the number of penalties called
team_penalties = df.groupby('Against')['Beneficiary'].count().reset_index()

# Sort the data in descending order to find the team with the most penalties
team_penalties = team_penalties.sort_values(by='Beneficiary', ascending=False)

# Print the top team
print(f"Team with the most penalties since 2009: {team_penalties.iloc[0]['Against']}")
```

This will give you the team New England that has been penalized the most during the 2009-2022 seasons.



## **Plot different features**

## 1.**BOXPLOT**

A boxplot is sometimes known as the box and whisker plot.
It shows the distribution of the quantitative data that
represents the comparisons between variables. boxplot shows
the quartiles of the dataset while the whiskers extend to
show the rest of the distribution i.e. the dots indicating
the presence of outliers. It is created using the
boxplot() method.

```python
sns.boxplot(x=df['Offside'], y=df['Total_Count'], data=df)
plt.show()
```

In this boxplot shows the offside count is high when it is 8

## 2.Histogram

A histogram is a graph that shows the frequency of numerical data using rectangles.

```python
df. Off_Yards.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Total_Count")
plt.ylabel("Total_Yards")
plt.xlabel(" Off_Yards");
```

The highest offyards is above 800 and the lowest offyard is below 200.

## 3.**Scatter Plot**

The scatter plot is a mainstay of statistical visualization.
It depicts the joint distribution of two variables using
a cloud of points, where each point represents an
observation in the dataset. This depiction allows the eye
to infer a substantial amount of information about whether
there is any meaningful relationship between them. It is
plotted using the scatterplot() method.

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['Off_Yards'], df['Offside'])
ax.set_xlabel('St_Count')
ax.set_ylabel('Off_Yards')
plt.show()
```

In this graph shows that most of the results are lies in between 0-60.

## 4.**Bar Plot**

A barplot is basically used to aggregate the categorical
data according  to some methods and by default its the
mean. It can also be understood  as a visualization of
the group by action. To use this plot we choose  a
categorical column for the x axis and a numerical column
for the y  axis and we see that it creates a plot taking
a mean per categorical  column. It can be created using
the barplot() method.

```python
sns.barplot(df)
```

From this bar plot shows the total offyards is above 50 and it is the highest frequent.

## 5.**Count Plot**

A countplot basically counts the categories and returns a
count of their occurrences. It is one of the most simple
plots provided by the seaborn library. It can be created
using the countplot() method.

```python
sns.countplot(x=df['Offside'], data=df)
plt.show()
```

The graph shows the highest offside count is 2500 and the lowest count is less than 500.

## **6.Lineplot**

For certain datasets, you may want to consider changes as a
function of time in one variable, or as a similarly
continuous variable. In this case, drawing a line-plot is
a better option. It is plotted using the lineplot() method.

```python
sns.lineplot(x=("Total_Yards"),y=("Total_Count"),data=df.head(20))
```

From this graph shows the highest total yards is 80 and the lowest total yards is above 20.
